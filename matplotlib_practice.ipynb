# Matplotlib example workflow

```python
# import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y=[22,55,99,66]

# 2. Setup plot
fig,ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Practice Plot",
      xlabel = "X axis",
      ylabel="Y axis")

# 5. Save & show (you save the graph)
fig.savefig("sample-plot.png")
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* subplot

```python
# Create some data
x = np.linspace(0,10,100)
x[:10]
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Use same data to make a scatter
fig,ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter":10,
                    "Peanut butter":8,
                    "Cashew butter":12}
fig,ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Nut butter store", 
      ylabel="Price($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots

```python
# Subplots option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                         ncols=2,
                                         figsize=(10,5))
# Plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(10,5))
# Plot to each different index
ax[0,0].plot(x,np.sin(x/2));
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
# Try on a different dataset
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Create a histogram of age
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.tail()
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
#Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([50,60])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# Customize the plot
ax.set(title="Heart Disease and Cholestreol Levels",
      xlabel="Age",
      ylabel="Cholestreol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize= (10,10),
                              sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])
# Customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
       ylabel="Cholestreol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
# Customize ax1
ax1.set(title='Heart disease and Max Heart Rate',
       xlabel="Age",
       ylabel='Max Heart Rate')

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid")
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                              ncols = 1,
                              figsize= (10,10),
                              sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")
# Customize ax0
ax0.set(title="Heart disease and Cholesterol Levels",
       ylabel="Cholestreol")
# Change the x axis range
ax0.set_xlim([50,80])
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

#Add data to ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")
# Customize ax1
ax1.set(title='Heart disease and Max Heart Rate',
       xlabel="Age",
       ylabel='Max Heart Rate')

# Add legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Change the x axis range
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
# Add a meanline
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle='--');

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
fig.savefig("analysis.png")
```
